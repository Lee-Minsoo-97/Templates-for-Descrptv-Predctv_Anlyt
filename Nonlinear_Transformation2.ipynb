```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Cost of Power.xlsx')
```

```python
data.head()
```

```python
plt.scatter(data['Units'], data['Cost'])
```

```python
model_CU = smf.ols('Cost~Units',data = data).fit()
print(model_CU.summary())
```

```python
fitted_values = model_CU.fittedvalues
residuals = model_CU.resid
plt.scatter(fitted_values, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
model_CUUsq = smf.ols('Cost~ Units + I(Units**2)',data = data).fit()
print(model_CUUsq.summary())
```

```python
fitted_values = model_CUUsq.fittedvalues
residuals = model_CUUsq.resid
plt.scatter(fitted_values, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
model_CLu = smf.ols('Cost~ np.log(Units)',data = data).fit()
print(model_CLu.summary())
```
